```python
#a function to get the properties of a large data set
def set_largedata(dataset):
    size = dataset.count()
    data_buffer={'last_buffer':None}
    classified_data = []
    if(size > 100000):
        buffer_size= size / 100000
    buffer_size_ceil = math.ceil(buffer_size)
    buffer_size_floor = buffer_size - math.floor(buffer_size)
    buffer_size_floor = round(buffer_size_floor,6)
    completed_data_chunk = math.floor(buffer_size) + buffer_size_floor
    data_chunks = 0
    
    while data_chunks < buffer_size_ceil:
        if((completed_data_chunk - data_chunks) >= 1):
            data_buffer[f'{data_chunks}'] = set(dataset[data_chunks * 100000 : (data_chunks * 100000) + 100001])
#             data_buffer[f'{data_chunks}'] = set(dataset[0:100+1])
            data_chunks += 1
                                           
        else:                                  
            data_buffer['last_buffer'] =set(dataset[data_chunks * 100000:round(completed_data_chunk * 100000) + 1])
            break
            
                                           
    
    for i in data_buffer:
#         print(data_buffer[i])
        for data in data_buffer[i]:
            classified_data.append(data)
        
    
    print(set(classified_data))
                                           
    
        
        
    
```

```python
#City Population Dataset

def convert_to_int(data):
    print(type(data))

def percentage(selectedSize,originSize):
    dataFrame={'percentage':[]}
    length = selectedSize.size
    for i in range(0,length):
        sample = selectedSize[i].replace(',','')
        total = originSize[i].replace(',','')
        percentage = round(int(sample)/int(total) * 100,2)
        dataFrame['percentage'].append(percentage)
    return dataFrame


population = city_dataset['Population']
users = city_dataset['Users']

# percentage(users,population)
# print(population.size)

city_percentage = percentage(city_dataset['Users'],city_dataset['Population'])
print(city_percentage)

percent_data = pd.DataFrame(city_percentage)
city_dataset.insert(1,'PercentageUsers',percent_data['percentage'])
# city_dataset.to_csv('C:/Users/Ekomzy/Desktop/Data Science/modified_datasets/city_to_customers')
new_city_dataset = city_dataset.sort_values('PercentageUsers',ascending=True)
new_city_dataset
```
